```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python

data=[]
for i in range(12594):
    img = Image.open('E:\dsa project\data\\'+str(i)+'.jpg')
    numpydata = np.asarray(img)
    numpydata=numpydata.reshape(-1)
    data.append(numpydata)
    if i%100==0 :
        print(i)
data=np.array(data)
print(data.shape)
from matplotlib import pyplot as plt
```

```python
mean=np.mean(data.T,axis=1)
normalised=data-mean
cov=np.cov(normalised.T)
ev,ex=np.linalg.eig(cov)
print("done")
```

```python
np.save('E:\dsa project\mean.npy',mean)
np.save('E:\dsa project\eigenValues.npy',ev)
np.save('E:\dsa project\eigenVecs.npy',ex)
```
